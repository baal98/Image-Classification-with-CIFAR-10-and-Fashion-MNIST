Now that we have defined our models, it's time to train them on our datasets. Training a model involves feeding it our training data and adjusting the model's weights based on the output error. Let's proceed with the training process for both datasets.


## Import dependencies

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model

# Load the trained models
cifar_model = load_model('../saved_models/cifar_model_enhanced.h5')
fmnist_model = load_model('../saved_models/fmnist_model_enhanced.h5')

data = np.load('../saved_models/cifar_data.npz')
train_images_cifar = data['train_images']
test_images_cifar = data['test_images']
train_labels_cifar = data['train_labels']
test_labels_cifar = data['test_labels']


data = np.load('../saved_models/fmnist_data.npz')
train_images_fmnist = data['train_images']
test_images_fmnist = data['test_images']
train_labels_fmnist = data['train_labels']
test_labels_fmnist = data['test_labels']
```

### Training the CIFAR-10 Model

We'll train the CIFAR-10 model using the training images and labels. Additionally, we'll use a portion of the training data as validation data. This helps monitor the model's performance on unseen data during training.


```python
# Reshaping the data to fit the model
train_images_cifar = train_images_cifar.reshape((-1, 32, 32, 3))

# Training the CIFAR-10 model
history_cifar = cifar_model.fit(train_images_cifar, train_labels_cifar, 
                                epochs=5, 
                                validation_split=0.2)
```

After training the CIFAR-10 model, it's a good practice to visualize the training and validation loss and accuracy. This helps in diagnosing issues like overfitting.

```python
# Visualizing training results for CIFAR-10 model
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history_cifar.history['accuracy'], label='Training Accuracy')
plt.plot(history_cifar.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history_cifar.history['loss'], label='Training Loss')
plt.plot(history_cifar.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

### Training the Fashion MNIST Model

Similar to CIFAR-10, we'll train the Fashion MNIST model using its respective training images and labels and validate its performance using a subset of the training data.


```python
# Reshaping the data to fit the model
train_images_fmnist = train_images_fmnist.reshape((-1, 28, 28, 1))

# Training the Fashion MNIST model
history_fmnist = fmnist_model.fit(train_images_fmnist, train_labels_fmnist, 
                                  epochs=4, 
                                  validation_split=0.2)
```

Again, after training, we'll visualize the training and validation loss and accuracy for the Fashion MNIST model.


```python
# Visualizing training results for Fashion MNIST model
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history_fmnist.history['accuracy'], label='Training Accuracy')
plt.plot(history_fmnist.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history_fmnist.history['loss'], label='Training Loss')
plt.plot(history_fmnist.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
